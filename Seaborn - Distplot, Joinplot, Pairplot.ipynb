```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

#### Co relation with Heatmap

a correlation heatmap ises colored cells, typically in a monochromatic scale, to show a 2D correlation matrix bet two discrete dimensions or event types. it is very important in feature selection

```python
df.corr()

# tip and bill are positively corelated
# meaning as the bill increases the tip also incerases
```

```python
# correlation will be cal only for int or float features
df.dtypes
```

```python
sns.heatmap(df.corr())
```

#### Joinplot

A join plt allows to study the relationship bet 2 numeric variables. the central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(x ='tip', y='total_bill', data=df, kind ='hex')
```

```python
sns.jointplot(x ='tip', y='total_bill', data=df, kind ='reg')
```

#### Pair plot

A pair plot is also known as scatterplot, in which one variable in the same data row is matched with another variables value, like this: Pair plots are just elaboration on showing all variables paired with all the other variables.


```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
```

#### Dist Plot
It helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins =10)
```

#### Categorical Plots

Seaborn also helps us to analyse the categorical data.

BoxpLot
violinplot
countplot
barplot

```python
# count plot
sns.countplot('sex', data=df)
```

```python
sns.countplot(y = 'sex', data=df)
```

```python
sns.countplot('day', data=df)
```

```python
# Bar PLot 
sns.barplot(x = 'total_bill', y='sex', data=df)
```

#### Box PLot

a box and whisker plot is a graph that present info from five-number summary

```python
sns.boxplot('sex', 'total_bill', data =df)
```

```python
sns.boxplot(x = 'day', y='total_bill', data =df, palette='rainbow')
```

```python
sns.boxplot(data = df, orient='v')
```

```python
sns.boxplot(x = 'total_bill', y='day', hue='smoker', data=df)
```

#### Violin Plot

violin plot helps us to see both the distribution of data in terms of kernel density estimation and the box plot

```python
sns.violinplot(x= 'total_bill', y='day', data=df, palette='rainbow')
```
